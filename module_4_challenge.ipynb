## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty 14,17,19
df["unit_price"] = df["unit_price"].astype("float")
df["qty"] = df["qty"].astype("float")
df = df.dropna()
df = df.drop_duplicates()
df['line_subtotal'] = df['unit_price']*df['qty']
df = df.sort_values('line_subtotal', ascending=False)
client_33651 = pd.DataFrame(df.loc[(df["client_id"] == 33615)])
client_66037 = pd.DataFrame(df.loc[(df["client_id"] == 66037)])
print(client_66037["line_subtotal"].sum())
print(client_33651["line_subtotal"].sum())
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
# "Apply" the function to the DataFrame
df['order_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['order_weight'].apply(lambda x: x*10 if x < 50 else x*7)
df = df.sort_values('line_subtotal', ascending=False)
df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['before_tax']= df['shipping_price'] + df['line_subtotal']
df['line_price'] = df['before_tax'].apply(lambda x: x*0.0925 ) + df['shipping_price'] + df['line_subtotal']
df_sort = df.sort_values('line_price', ascending=False)
df_sort.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price'] 
df_sort = df.sort_values('line_cost', ascending=False)
df_sort.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['line_profit'] = df['line_price'] - df['line_cost'] 
df_sort = df.sort_values('line_profit', ascending=False)
df_sort.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
for x in clientidlist_df:
    x_spend = df.loc[(df["client_id"] == x)]
    print (x ,x_spend["line_price"].sum())
    
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

Total_unit_purchased = []
Total_shipping_price = []
Total_revenue = []
Total_profit = []
client_id=[]
for x in clientidlist_df:
    client_id.append(x)
    x_spend = df.loc[(df["client_id"] == x)]
    Total_unit_purchased.append(x_spend["qty"].sum())
    Total_shipping_price.append(x_spend["shipping_price"].sum())
    Total_revenue.append(x_spend["line_price"].sum())
    Total_profit.append(x_spend["line_profit"].sum())

list_of_tuples = list(zip(client_id, Total_unit_purchased,Total_shipping_price,Total_revenue,Total_profit)) 
# Assign data to tuples.
list_of_tuples
Summary_df = pd.DataFrame(list_of_tuples,
                  columns=['Client_id', 'Total unit purchased','Total shipping price','Total revenue','Total profit'])

Summary_df = pd.DataFrame(Summary_df.sort_values('Total profit', ascending=False))
# Print data.
Summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
Summary_df_M=Summary_df
tup_of_dollars = Summary_df["Total shipping price"]/1000000
Summary_df["Total shipping price"] = tup_of_dollars.round(2)
Summary_df_M["Total shipping price"]= '$'+ Summary_df["Total shipping price"].astype(str) + 'M'
rup_of_dollars = Summary_df["Total revenue"]/1000000
Summary_df["Total revenue"] = rup_of_dollars.round(2)
Summary_df_M["Total revenue"]= '$'+Summary_df["Total revenue"].astype(str) + 'M'
tpp_of_dollars = Summary_df["Total profit"]/1000000
Summary_df["Total profit"] = tpp_of_dollars.round(2)
Summary_df_M["Total profit"]= '$'+Summary_df["Total profit"].astype(str) + 'M'


Summary_df_M
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
Summary_df = pd.DataFrame(Summary_df.sort_values('Total profit', ascending=False))

Summary_df
```
